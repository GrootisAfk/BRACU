```python
#Task 01

class Student:
    c: object
    a: object
    d: object
    n: object
    
    ID = 0

    def __init__(self, name, dept, age, cgpa) -> object:
        self.n = name
        self.d = dept
        self.a = age
        self.c = cgpa
        Student.ID += 1

    def get_details(self):
        print(f"ID: {Student.ID}")
        print(f"Name: {self.n}")
        print(f"Department: {self.d}")
        print(f"Age: {self.a}")
        print(f"CGPA: {self.c}")

    @classmethod
    def from_String(cls, info: object):
        name, dept, age, cgpa = info.split("-")
        x = cls(name, dept, age, cgpa)
        return x


s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
```

```python
#Task 02

class Assassin:
    x: int = 0

    def __init__(self, name, rate) -> object:
        self.y = name
        self.z = rate
        Assassin.x += 1

    @classmethod
    def failureRate(cls, name, fail: object):
        return cls(name, 100 - fail)
        Assassin.x = Assassin.x + 1

    @classmethod
    def failurePercentage(cls, name, percentage: object):
        return cls(name, 100 - percentage)
        Assassin.x = Assassin.x + 1

    def printDetails(self):
        print("Name: {}".format(self.y))
        print("Success rate: {}%".format(self.z))
        print(f"Total number of Assassin: {self.x}")


john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate('Nagisa', 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task 03

class Passenger:
    count = 0

    def __init__(self, name, taka=450):
        self.name = name
        self.taka = taka
        Passenger.count += 1

    def set_bag_weight(self, weight):
        self.weight = weight
        if self.weight <= 20:
            self.taka = self.taka
        elif self.weight > 21 and self.weight < 50:
            self.taka = self.taka + 50
        elif self.weight > 50:
            self.taka = self.taka + 100

    def printDetail(self):
        print('Name:', self.name)
        print('Bus Fare:', self.taka, 'taka')


# ====================================================================
print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
# task 05

import datetime


class Employee:

    workingPeriod: object
    name: object

    def __init__(self, x, y: object):

        self.name = x

        self.workingPeriod = y

    @classmethod
    def employeeByJoiningYear(cls, x, y) -> object:
        return cls(x, datetime.date.today().year - y)


def experienceCheck(y, gender):

    if y >= 3:
        employee_experience = " is experienced"
    else:
        employee_experience = " is not experienced"

    if gender.upper() != "FEMALE":
        return "He" + employee_experience
    else:

        return "She" + employee_experience


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(experienceCheck(2, 'male'))
print(experienceCheck(3, 'female'))
```

```python
# Task 04

class Travel:
    count = 0

    def __init__(self, source, destination):
        self.source = source
        self.destination = destination
        self.time = 1
        Travel.count += 1

    def set_time(self, time):
        self.time = time

    def set_destination(self, another_destination):
        self.destination = another_destination

    def set_source(self, other_source):
        self.source = other_source

    def display_travel_info(self):
        return 'Source:{}\nDestination:{}\nFlight Time:{}:00'.format(self.source, self.destination, self.time)


print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka", "India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur", "Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka", "New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka", "India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
#Task 07

class Cat:
    
    Number_of_cats=0
    
    def __init__(self,color,action):
        self.color=color
        self.action=action
        Cat.Number_of_cats+=1
        
    @classmethod    
    
    def no_parameter(cls):
        color,action='White','sitting'
        objective=cls(color,action)
        return objective
    
    @classmethod
    
    def first_parameter(cls,info):
        color,action=info,'sitting'
        objective1=cls(color,action)
        return objective1
    
    @classmethod
    
    def second_parameter(cls,new):
        color,action="Grey",new
        objective2=cls(color,action)
        return objective2
    
    def printCat(self):
        print(self.color,("cat is"),self.action)
        
    def changeColor(self,new_color):
        self.color=new_color
        
#================================================================
print("Total number of cats:", Cat.Number_of_cats) 
c1 = Cat.no_parameter() 
c2 = Cat.first_parameter("Black") 
c3 = Cat("Brown", "jumping") 
c4 = Cat("Red", "purring") 
c5 = Cat.second_parameter("playing")
print("=======================") 
c1.printCat() 
c2.printCat() 
c3.printCat() 
c4.printCat() 
c5.printCat() 
c1.changeColor("Blue") 
c3.changeColor("Purple") 
c1.printCat() 
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```
