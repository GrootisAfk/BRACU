```python
# Question 01

class Calculator:
    def __init__(self):
        self.x = 0
        self.y = 0
        print("Let's Calculate!")

    def add(self, x, y):
        self.x += x
        self.y += y
        print('Value 1: ', x)
        print('Operator: ', '+')
        print('Value 2: ', y)
        print('Result: ', x + y)

    def multiply(self, x, y):
        self.x += x
        self.y += y
        print('Value 1: ', x)
        print('Operator: ', '*')
        print('Value 2: ', y)
        print('Result: ', x * y)

    def divide(self, x, y):
        self.x += x
        self.y += y
        print('Value 1: ', x)
        print('Operator: ', '/')
        print('Value 2: ', y)
        print('Result: ', x / y)

    def subtract(self, x, y):
        self.x += x
        self.y += y
        print('Value 1: ', x)
        print('Operator: ', '-')
        print('Value 2: ', y)
        if x < y:
            print('Result: ', y - x)
        else:
            print('Result: ', x - y)


z = Calculator()
val1 = int(input("Enter a value:"))
operator = str(input("Enter an operator:"))
val2 = int(input("Enter another value:"))
if operator in '+':
    z.add(val1, val2)
elif operator in '*':
    z.multiply(val1, val2)
elif operator in '/':
    z.divide(val1, val2)
elif operator in '-':
    z.subtract(val1, val2)
```

```python
# Question 02

class Customer:
    def __init__(self, x):
        self.name = x

    def greet(self, name=None):
        if name != None:
            print('Hello', self.name + '!')
        else:
            print('Hello!')

    def purchase(self, *y):
        print(self.name + ',', 'you have purchased', len(y), 'item(s):')
        for i in y:
            print(i)


customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
# Question 03

class Panda:
    def __init__(self, x, y, z):
        self.name = x
        self.gender = y
        self.age = z

    def sleep(self, a: object = ""):
        if a == "":
            return "{}'s duration is unknown thus should have only bamboo leaves".format(self.name)
        elif int(a) >= int(3) and int(a) <= int(5):
            return "{} sleeps {} hours daily and should have Mixed Veggies".format(self.name, a)
        elif int(a) >= int(6) and int(a) <= int(8):
            return "{} sleeps {} hours daily and should have Eggplant & Tofu".format(self.name, a)
        elif int(a) >= int(9) and int(a) <= int(11):
            return "{} sleeps {} hours daily and should have Broccoli Chicken".format(self.name, a)


panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female", 3)
panda3 = Panda("Ming Ming", "Female", 8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name, panda1.gender, panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name, panda2.gender, panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name, panda3.gender, panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
# Question 04

class Cat:
    def __init__(self, color='White', action: object = 'sitting') -> object:
        self.x = color
        self.y = action

    def printCat(self, color="", action=""):
        if action == None or color is None:
            return
        print(self.x, 'cat is', self.y)

    def changeColor(self, color: object):
        self.x = color


c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
# Question 05

class Student:
    y: float

    def __init__(self, name: object = '') -> object:
        self.x = name
        self.y = 0.0

    def quizcalc(self, q1=0, q2=0, q3=0):
        self.y = (q1 + q2 + q3) / 3.0

    def printdetail(z):
        print("Hello default student" + z.x)
        print("Your Average score is ", z.y)


s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
print('--------------------------------')
s3.printdetail()
```

```python
#Question 06

class Vehicle:
    def __init__(self, x=0, y=0) -> object:
        self.x = x
        self.y = y

    def print_position(self):
        print("(" + str(self.x) + "," + str(self.y) + ")")

    def moveUp(self):
        self.y = self.y + 1

    def moveDown(self):
        self.y = self.y - 1

    def moveRight(self):
        self.x = self.x + 1

    def moveLeft(self):
        self.x = self.x - 1

car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
# Question 07

class Programmer:
    count: int
    z: object

    def __init__(self, name, programming: object, experience):
        self.x = name
        self.y = programming
        self.z = experience
        self.count = 0

    def printDetails(self):
        if self.count == 0:
            print('Horray! A new programmer is born')
        else:
            print('Updating experience of ' + self.x)
        print('Name:', self.x)
        print('Programming:', self.y)
        print('Experience:', self.z, 'years')

    def addExp(self, a: object):
        self.z += a
        self.count += 1


p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()
```

```python
# Question 08

class Student:
    z: str
    y: object
    x: object

    def __init__(self, name, ID, department='CSE') -> object:
        self.x = name
        self.y = ID
        self.z = department

    def dailyEffort(self, n: object):
        self.time = n

    def printDetails(self):
        print('Name:', self.x)
        print('ID:', self.y)
        print('Department:', self.z)
        print('Daily Effort:', self.time, 'hour(s)')
        if self.time <= 2:
            print('Suggestion: Should give more effort!')
        elif self.time <= 4:
            print('Suggestion: Keep up the good work!')
        else:
            print('Suggestion: Excellent! Now motivate others.')


harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
# Question 09

class Patient:
    x: object
    y: object

    def __init__(self, name, age):
        self.x = name
        self.y = age

    def add_Symptom(self, *a: object) -> object:
        for n in range(0, len(a) + 1):
            if len(a) == 1:
                self.sym = a[0]
            elif len(a) == 2:
                self.sym = a[0] + ',' + a[1]
            else:
                self.sym = a[0] + ',' + a[1] + ',' + a[2]

    def printPatientDetail(self):
        assert isinstance(self.x, object)
        print('Name:', self.x)
        print('Age:', self.y)
        print('Symptoms:', self.sym)


p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```python
# Question 11

class Shinobi:
    m: int
    z: int
    x: object
    y: object

    def __init__(self, name, rank):
        self.x = name
        self.y = rank
        self.z = 0
        self.m = 0

    def calSalary(self, number: object):
        self.m = number
        if self.y == "Genin":
            self.z = number * 50
        elif self.y == "Chunin":
            self.z = number * 100
        else:
            self.z = number * 500

    def changeRank(self, rank):
        self.y = rank

    def printInfo(self) -> object:
        print('Name:', self.x, '\nRank:', self.y, 'nNumber of missions:', self.m, '\nSalary:', self.z)


naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```

```python
#Question 12

class ParcelKoro:
    def __init__(self, name = 'No name set', weight = 0):
        self.x = name
        self.y = weight
        self.z = 0

    def calculateFee(self, i: object = ''):
        if i == '':
            j = 50
        else:
            j: int = 100
        if self.y == 0:
            self.z = 0
        else:
            self.z = (self.y * 20) + j

    def printDetails(self):
        print('Customer Name:', self.x)
        print('Product Weight:', self.y)
        print('Total fee:', self.z)
        print('*******************')


print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.y = 15
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
#Question 13

class Batsman:
    def __init__(self, *a):
        if isinstance(a[0],
                      int):
            self.x = "New Batsman"
            self.y = a[0]
            self.z = a[1]
        elif isinstance(a[0], str):
            self.x = a[0]
            self.y = a[1]
            self.z = a[2]

    def battingStrikeRate(self):
        return (self.y / self.z) * 100

    def setName(self, name):
        self.x = name

    def printCareerStatistics(self):
        print("Name:", self.x)
        print("Runs Scored:", self.y, ", Balls Faced:", self.z)


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
# Question 14

class EPL_Team:
    football = 0

    def __init__(self, Name, Song=None):
        self.x = Name
        self.y = Song
        if (Song == None):
            self.y = "No Slogan"

    def changeSong(self, z: object):
        self.y = z

    def increaseTitle(self):
        self.football += 1

    @property
    def showClubInfo(self):
        return "Name : {}\nSong: {}\nTotal No. of Title: {}".format(self.x, self.y, self.football)


manu = EPL_Team("Manchester United", "Glory Glory Man United")
chelsea = EPL_Team("Chelsea")
print("===================")
print(manu.showClubInfo)
print("##################")
manu.increaseTitle()
print(manu.showClubInfo)
print("===================")
print(chelsea.showClubInfo)
chelsea.changeSong('Keep the blue flag flying high')
print(chelsea.showClubInfo)
```

```python
#Question 16

class Author:
    def __init__(self, Name: object = "Default", *Books: object):
        self.x = Name
        self.y = []
        z: object
        for z in Books:
            self.y.append(z)

    def addBooks(self, *books):
        for book in books:
            self.y.append(book)

    def printDetails(self):
        print("Author Name:", self.x)
        print("-------")
        print("List of Books:")
        for book in self.y:
            print(book)

    def changeName(self, name):
        self.x = name


auth1 = Author('Humayun Ahmed')
auth1.addBooks('Deyal', 'Megher Opor Bari')
auth1.printDetails()
print("===================")
auth2 = Author()
print(auth2.x)
auth2.changeName('Mario Puzo')
auth2.addBooks('The Godfather', 'Omerta', 'The Sicilian')
print("===================")
auth2.printDetails()
print("===================")
auth3 = Author('Paolo Coelho', 'The Alchemist', 'The Fifth Mountain')
auth3.printDetails()
```
