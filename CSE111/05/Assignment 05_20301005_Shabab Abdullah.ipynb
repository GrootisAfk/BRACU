```python
#Question 01

class Marks:
    mark = []

    def __init__(self, x) -> object:
        self.mark = x
        Marks.mark.append(self.mark)

    def __add__(self, y):
        z: Marks = Marks(self.mark + y.mark)
        return z


Q1 = Marks(int(input("Quiz 1 (out of 10): ")))
Q2 = Marks(int(input("Quiz 2 (out of 10): ")))
Lab = Marks(int(input("Lab (out of 30): ")))
Mid = Marks(int(input("Mid (out of 20): ")))
Final = Marks(int(input("Final (out of 30): ")))
total = Q1 + Q2 + Lab + Mid + Final
print("Total marks: {}".format(total.mark))
```

```python
#Question 02

class Teacher:
    def __init__(self, Name, Dept: object) -> object:
        self.n = Name
        self.d = Dept
        self.lst = []

    def addCourse(self, course: object):
        do_add = course.getCourse()
        self.lst.append(do_add)

    def printDetail(self):
        print("====================================")
        print(f"Name: {self.n}")
        print(f"Department: {self.d}")
        print("List of courses")
        print("====================================")
        for i in self.lst:
            print(i)
        print("====================================")


class Course:
    x: object

    def __init__(self, name_of_the_quiz: object):
        self.x = name_of_the_quiz

    def getCourse(self) -> object:
        return self.x

t1 = Teacher("Saad Abdullah", "CSE")
t2 = Teacher("Mumit Khan", "CSE")
t3 = Teacher("Sadia Kazi", "CSE")
c1 = Course("CSE 110 Programming Language I")
c2 = Course("CSE 111 Programming Language-II")
c3 = Course("CSE 220 Data Structures")
c4 = Course("CSE 221 Algorithms")
c5 = Course("CCSE 230 Discrete Mathematics")
c6 = Course("CSE 310 Object Oriented Programming")
c7 = Course("CSE 320 Data Communications")
c8 = Course("CSE 340 Computer Architecture")
t1.addCourse(c1)
t1.addCourse(c2)
t2.addCourse(c3)
t2.addCourse(c4)
t2.addCourse(c5)
t3.addCourse(c6)
t3.addCourse(c7)
t3.addCourse(c8)
t1.printDetail()
t2.printDetail()
t3.printDetail()
```

```python
#Question 03

class Team:
    def __init__(self, country: object = None):
        self._c = country
        self.x = []

    def setName(self, country) -> object:
        self._c = country

    def addPlayer(self, add: object):
        self.x.append(add.name)

    def printDetail(self):
        print('=====================')
        print('Team:', self._c)
        print('List of Players:')
        print(self.x)
        print('=====================')


class Player:
    def __init__(self, name):
        self.name = name


b = Team()
b.setName('Bangladesh')
mashrafi = Player("Mashrafi")
b.addPlayer(mashrafi)
tamim = Player("Tamim")
b.addPlayer(tamim)
b.printDetail()
a = Team("Australia")
ponting = Player("Ponting")
a.addPlayer(ponting)
lee = Player("Lee")
a.addPlayer(lee)
a.printDetail()
```

```python
#Question 04

class Color:
    def __init__(self, colour: object) -> object:
        self.x = colour

    def __add__(self, other: object):
        C3 = self.x + other.x

        if 'red' in C3 and 'blue' in C3:
            return Color('Violet')
        elif "red" in C3 and 'yellow' in C3:
            return Color('Orange')
        elif 'yellow' in C3 and 'BLUE' in C3:
            return Color('Green')


C1 = Color(input("First Color: ").lower())
C2 = Color(input("Second Color: ").lower())
C3 = C1 + C2
print("Color formed:", C3.x)
```

```python
#Question 05

import math

class Circle:
    __radius: object

    def __init__(self, radius):
        self.__radius = radius

    def getRadius(self):
        return self.__radius

    def setRadius(self, new_radius):
        self.__radius = new_radius

    @property
    def area(self):
        area = math.pi * self.__radius ** 2
        return area

    def __add__(self, other):
        total = self.__radius + other.__radius
        return Circle(total)


c1 = Circle(4)
print("First circle radius:", c1.getRadius())
print("First circle area:", c1.area)

c2 = Circle(5)
print("Second circle radius:", c2.getRadius())
print("Second circle area:", c2.area)

c3 = c1 + c2
print("Third circle radius:", c3.getRadius())
print("Third circle area:", c3.area)
```

```python
#Question 06

class Triangle:
    x: object
    y: object

    def __init__(self, base, height):
        self.x = base
        self.y = height

    @property
    def getBase(self):
        return self.x

    def getHeight(self):
        return self.y

    def setBase(self, base_2):
        self.x = base_2

    def setHeight(self, height_2: object):
        self.y = height_2

    def area(self):
        area = 0.5 * self.x * self.y
        return area

    def __sub__(self, other):
        base_subtract = self.x - other.x
        height_subtract = self.y - other.y
        return Triangle(base_subtract, height_subtract)


t1 = Triangle(10, 5)
print("First Triangle Base:", t1.getBase)
print("First Triangle Height:", t1.getHeight())
print("First Triangle area:", t1.area())

t2 = Triangle(5, 3)
print("Second Triangle Base:", t2.getBase)
print("Second Triangle Height:", t2.getHeight())
print("Second Triangle area:", t2.area())

t3 = t1 - t2
print("Third Triangle Base:", t3.getBase)
print("Third Triangle Height:", t3.getHeight())
print("Third Triangle area:", t3.area())
```

```python
#Question 08

class Coordinates:
    def __init__(self, x_coordinate, y_coordinate) -> object:
        self.x_coordinate = x_coordinate
        self.y_coordinate = y_coordinate

    def __sub__(self, other):
        do_subtract_x = self.x_coordinate - other.x_coordinate
        do_subtract_y = self.y_coordinate - other.y_coordinate
        return Coordinates(do_subtract_x, do_subtract_y)

    def __mul__(self, other):
        do_multi_x = self.x_coordinate * other.x_coordinate
        do_multi_y = self.y_coordinate * other.y_coordinate
        return Coordinates(do_multi_x, do_multi_y)

    def detail(self):
        return f"({self.x_coordinate}, {self.y_coordinate})"

    def __eq__(self, other):
        if self.x_coordinate == other.x_coordinate and self.y_coordinate == other.y_coordinate:
            return "The calculated coordinates are the same."
        else:
            return "The calculated coordinates are NOT the same."


p1 = Coordinates(int(input()), int(input()))
p2 = Coordinates(int(input()), int(input()))

p4 = p1 - p2
print(p4.detail())

p5 = p1 * p2
print(p5.detail())

point_check = (p4 == p5)
print(point_check)
```
